```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt

# Cargar el archivo CSV en un DataFrame de pandas
df = pd.read_csv("../results/Bakery sales ingredientes por dia.csv")

# Mostrar las columnas del DataFrame
print(df.columns)
print(df.head())
```

```python
# Establecer la columna 'ingrediente' como índice y transponer el DataFrame
# es decir, ahora cada columna es una serie de un ingrediente
df_ingredientes = df.set_index('ingrediente').T
```

```python
for ingrediente in df_ingredientes.columns:
    # Obtener la serie de tiempo para el ingrediente
    serie = df_ingredientes[ingrediente]

    # Ajuste con SimpleExpSmoothing
    # Se elige el alfa (smoothing_level) óptimo automaticamente 
    fit_ses = SimpleExpSmoothing(serie).fit()
    forecast_ses = fit_ses.fittedvalues

    # Ajuste con Holt
    # Alfa y Beta óptimos tambien se eligen automaticamente
    fit_holt = Holt(serie).fit()
    forecast_holt = fit_holt.fittedvalues

    # Graficar la serie de datos y los ajustes
    plt.figure(figsize=(10, 6))
    plt.plot(serie, label='Datos originales')
    plt.plot(forecast_ses, label='SimpleExpSmoothing')
    plt.plot(forecast_holt, label='Holt')
    plt.title(f'Ingrediente: {ingrediente}')
    plt.legend()
    plt.show()
```
