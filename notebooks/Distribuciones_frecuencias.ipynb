```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Cargar los datos
data = pd.read_csv('ingredientes_diarios_agrupados.csv', parse_dates=['date'], index_col='date')

# Preprocesamiento de datos
data.index = pd.to_datetime(data.index)
data = data.asfreq('D')
data = data.fillna(method='ffill')  # Rellenar valores faltantes

# Seleccionamos la columna de interés
ingrediente = 'SAL'
series = data[ingrediente]

# Graficar la serie temporal
plt.figure(figsize=(10,6))
plt.plot(series, label='Datos históricos', color='blue')
plt.title(f'Serie temporal de {ingrediente}')
plt.xlabel('Fecha')
plt.ylabel('Cantidad')
plt.legend()
plt.show()

# Identificar los parámetros p, d, q (usar ACF y PACF para esto)
plot_acf(series)
plot_pacf(series)
plt.show()

# Ajustar el modelo ARIMA
# Ejemplo de parámetros (p, d, q) -> puedes ajustarlos según el análisis de ACF y PACF
model = ARIMA(series, order=(5,1,0))  # (p, d, q)
model_fit = model.fit()

# Resumen del modelo ajustado
print(model_fit.summary())

# Realizar predicciones para el próximo año (365 días)
forecast = model_fit.forecast(steps=365)

# Graficar predicciones
plt.figure(figsize=(10,6))
plt.plot(series, label='Datos históricos', color='blue')
plt.plot(pd.date_range(series.index[-1], periods=365, freq='D'), forecast, label='Predicciones', color='red', linestyle='--')
plt.title(f'Predicciones para {ingrediente}')
plt.xlabel('Fecha')
plt.ylabel('Cantidad')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Leer el archivo CSV
file_path = r"Z:\Descargas\io-abc-xyz-main\data\Bakery sales ingredientes - demanda diaria pronostico.csv"  # Ruta al archivo
df = pd.read_csv(file_path)

# Agrupar los datos por el nombre del ingrediente (columna 'unique_id')
grouped = df.groupby('unique_id')

# Para cada grupo (ingrediente), calcular la distribución de frecuencias y graficarla
for ingredient, group in grouped:
    # Extraer los valores de la columna 'y' que quieres analizar
    y_values = group['y'].dropna().values  # Elimina valores nulos si los hay
    
    if len(y_values) > 0:
        # Ajuste de una distribución log-normal
        shape, loc, scale = stats.lognorm.fit(y_values, floc=0)  # floc=0 para forzar el parámetro de ubicación a 0
        
        # Calcular el ancho de los bins del histograma
        bin_edges = np.histogram_bin_edges(y_values, bins=15)
        bin_width = bin_edges[1] - bin_edges[0]
        
        # Graficar la distribución ajustada
        x = np.linspace(min(y_values), max(y_values), 100)
        p = stats.lognorm.pdf(x, shape, loc, scale)
        
        # Escalar la PDF para que coincida con la densidad del histograma
        p_scaled = p * len(y_values) * bin_width

        # Graficar los datos y la distribución ajustada
        plt.plot(x, p_scaled, 'k', linewidth=2, label="Log-Normal ajustada")
        plt.hist(y_values, bins=15, density=False, alpha=0.6, color='g', label="Histograma de los datos (frecuencia)")
        plt.title(f"Ajuste de la distribución Log-Normal para {ingredient}")
        plt.xlabel("Valores de y")
        plt.ylabel("Frecuencia")
        plt.legend()
        plt.show()

        # Realizar la prueba de Kolmogorov-Smirnov para log-normal
        ks_stat, ks_p_value = stats.kstest(y_values, 'lognorm', args=(shape, loc, scale))
        print(f"Estadístico KS para distribución Log-Normal de {ingredient}: {ks_stat}, p-valor: {ks_p_value}")

        # Transformar los datos a escala logarítmica
        log_y_values = np.log(y_values)

        # Realizar la prueba Anderson-Darling para distribución normal sobre los datos transformados
        ad_stat, critical_values, significance_level = stats.anderson(log_y_values, dist='norm')
        print(f"Estadístico Anderson-Darling para {ingredient}: {ad_stat}")
        print(f"Valores críticos: {critical_values}")
        
        # Evaluar el ajuste según Anderson-Darling
        critical_value_5_percent = critical_values[2]  # Valor crítico para 5% de significancia
        if ad_stat > critical_value_5_percent:
            print(f"Mal ajuste: El estadístico AD ({ad_stat}) supera el crítico de 5% ({critical_value_5_percent})")
        else:
            print(f"Buen ajuste: El estadístico AD ({ad_stat}) es menor que el crítico de 5% ({critical_value_5_percent})")

        # Obtener el cuantil de la distribución log-normal para una probabilidad específica (por ejemplo, 0.95)
        probabilidad = 0.95
        cuantil = stats.lognorm.ppf(probabilidad, shape, loc, scale)
        print(f"shape: {shape}, loc: {loc}, scale: {scale}")
        print(f"Cuantil de la distribución Log-Normal para {ingredient} con probabilidad {probabilidad}: {cuantil}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Leer el archivo CSV
file_path = r"Z:\Descargas\io-abc-xyz-main\data\Bakery sales ingredientes - demanda diaria pronostico.csv"  # Ruta al archivo
df = pd.read_csv(file_path)

# Agrupar los datos por el nombre del ingrediente (columna 'unique_id')
grouped = df.groupby('unique_id')

# Función para calcular el estadístico Anderson-Darling manualmente para una distribución Gamma
def anderson_darling_gamma(data, shape, loc, scale):
    n = len(data)
    sorted_data = np.sort(data)
    cdf_values = stats.gamma.cdf(sorted_data, shape, loc, scale)
    cdf_values = np.clip(cdf_values, 1e-6, 1 - 1e-6)  # Evitar log(0) o log(1)
    ad_stat = -n - np.sum((2 * np.arange(1, n + 1) - 1) / n * (np.log(cdf_values) + np.log(1 - cdf_values[::-1])))
    return ad_stat

# Para cada grupo (ingrediente), calcular la distribución de frecuencias y graficarla
for ingredient, group in grouped:
    # Extraer los valores de la columna 'y' que quieres analizar
    y_values = group['y'].dropna().values  # Elimina valores nulos si los hay
    
    if len(y_values) > 0:
        # Ajuste de una distribución Gamma sin fijar el parámetro de localización
        shape, loc, scale = stats.gamma.fit(y_values)  # No fijamos floc=0
        
        # Calcular el ancho de los bins del histograma
        bin_edges = np.histogram_bin_edges(y_values, bins=15)
        bin_width = bin_edges[1] - bin_edges[0]
        
        # Graficar la distribución ajustada
        x = np.linspace(min(y_values), max(y_values), 100)
        p = stats.gamma.pdf(x, shape, loc, scale)
        
        # Escalar la PDF para que coincida con la frecuencia del histograma
        p_scaled = p * len(y_values) * bin_width

        # Graficar los datos y la distribución ajustada
        plt.plot(x, p_scaled, 'k', linewidth=2, label="Gamma ajustada")
        plt.hist(y_values, bins=15, density=False, alpha=0.6, color='g', label="Histograma de los datos (frecuencia)")
        plt.title(f"Ajuste de la distribución Gamma para {ingredient}")
        plt.xlabel("Valores de y")
        plt.ylabel("Frecuencia")
        plt.legend()
        plt.show()

        # Realizar la prueba de Kolmogorov-Smirnov para la distribución Gamma
        ks_stat, ks_p_value = stats.kstest(y_values, 'gamma', args=(shape, loc, scale))
        print(f"Estadístico KS para distribución Gamma de {ingredient}: {ks_stat}, p-valor: {ks_p_value}")

        # Realizar el análisis de Anderson-Darling
        ad_stat = anderson_darling_gamma(y_values, shape, loc, scale)
        print(f"Estadístico Anderson-Darling para {ingredient}: {ad_stat}")

        # Comparar con un valor crítico aproximado (a ajustar según muestras o tablas específicas)
        critical_value = 1.061  # Para muestras grandes, nivel de significancia del 5%
        if ad_stat > critical_value:
            print(f"Mal ajuste: El estadístico AD ({ad_stat}) supera el crítico ({critical_value})")
        else:
            print(f"Buen ajuste: El estadístico AD ({ad_stat}) es menor que el crítico ({critical_value})")

        # Obtener el cuantil de la distribución Gamma para una probabilidad específica (por ejemplo, 0.95)
        probabilidad = 0.95
        cuantil = stats.gamma.ppf(probabilidad, shape, loc, scale)
        print(f"shape: {shape}, loc: {loc}, scale: {scale}")
        print(f"Cuantil de la distribución Gamma para {ingredient} con probabilidad {probabilidad}: {cuantil}")
```

```python
#SR = cuantil + dmin*LT       (cuantil = stats.gamma.ppf(probabilidad_servicio, alpha, loc, beta))
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

# Leer el archivo CSV
file_path = r"Z:\Descargas\io-abc-xyz-main\data\Bakery sales ingredientes - demanda diaria pronostico.csv"  # Ruta al archivo
df = pd.read_csv(file_path)

# Asegúrate de que la columna 'y' (demanda diaria) esté en formato numérico y no tenga valores nulos
df['y'] = pd.to_numeric(df['y'], errors='coerce')
df = df.dropna(subset=['y'])

# Agrupar los datos por el nombre del ingrediente
grouped = df.groupby('unique_id')  # Asumiendo que tienes una columna 'unique_id' para cada ingrediente

# Para cada grupo (ingrediente), calcular la distribución de frecuencias y graficarla
for ingredient, group in grouped:
    # Extraer los valores de la columna 'y' (demanda diaria)
    y_values = group['y'].values

    # Ajustar la distribución Gamma a los datos
    alpha, loc, beta = stats.gamma.fit(y_values)  # Estimar los parámetros de la distribución Gamma

    # Recalcular los valores de la distribución
    mean_gamma = alpha * beta
    std_gamma = np.sqrt(alpha * (beta**2))

    # Nivel de servicio deseado 
    probabilidad_servicio = 0.99

    # Calcular el cuantil para el nivel de servicio
    cuantil = stats.gamma.ppf(probabilidad_servicio, alpha, loc, beta) / 1000  # En kilogramos

    # Cálculo del tamaño de lote a pedir (q) usando la fórmula EOQ
    # Supongamos que:
    # D = demanda anual (se puede calcular como la suma de la demanda diaria multiplicada por 365 días)
    # K = costo por pedido
    # C1 = costo por mantener inventario
    D = np.sum(y_values) * (365 / 176) / 1000  # Demanda anual estimada en kg
    K = 50000  # Costo por pedido (ajustable según el caso)
    C1 = 20000  # Costo de mantener inventario por unidad (ajustable según el caso)

    q = np.sqrt((2 * D * K) / C1)  # Tamaño de lote usando EOQ (en kg)

    # Calcular el Punto de Reabastecimiento (ROP)
    lead_time = 7  # Tiempo de entrega (lead time) en días
    demanda_minima = np.min(y_values) / 1000  # Demanda mínima diaria observada (en kg)
    rop = cuantil + (demanda_minima * lead_time)  # ROP ajustado con el cuantil y el lead time

    # Imprimir los resultados para cada ingrediente
    print(f"Ajuste para {ingredient}:")
    print(f"Media: {mean_gamma / 1000:.2f} Kg")
    print(f"Desviación estándar: {std_gamma / 1000:.2f} Kg")
    print(f"Demanda diaria promedio: {np.mean(y_values) / 1000:.2f} Kg")
    print(f"Demanda mínima diaria: {demanda_minima:.2f} Kg")
    print(f"Demanda anual: {D:.2f} Kg")
    print(f"Cuantil para el nivel de servicio de {probabilidad_servicio * 100}%: {cuantil:.2f} Kg")
    print(f"Tamaño de lote a pedir (q): {q:.2f} Kg")  # Tamaño de lote en kg
    print(f"Punto de Reabastecimiento (ROP): {rop:.2f} Kg")  # ROP en kg
    print("-" * 40)

    # Graficar la distribución gamma ajustada y la frecuencia
    x = np.linspace(min(y_values), max(y_values), 1000)  # Generar un rango de valores para la distribución ajustada
    p = stats.gamma.pdf(x, alpha, loc, beta)  # Obtener la función de densidad de la distribución ajustada

    plt.plot(x, p, 'k', linewidth=2, label=f"Distribución Gamma ajustada ({ingredient})")  # Línea de la distribución Gamma ajustada
    plt.hist(y_values, bins=30, density=True, alpha=0.6, color='g', label=f"Histograma de los datos ({ingredient})")  # Histograma de los datos reales

    plt.title(f"Distribución Gamma ajustada para {ingredient}")
    plt.xlabel("Valores de demanda")
    plt.ylabel("Frecuencia")
    plt.legend()
    plt.show()
```
